```python
import pandas as pd
import numpy as np
```

```python
import certifi
import ssl

ssl_context = ssl.create_default_context(cafile=certifi.where())
```

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as ureq
my_url = "https://www.flipkart.com/womens-footwear/~sports-casual-shoes/pr?sid=osp%2Ciko&otracker=nmenu_sub_Women_0_Shoes&page=1"
openFile = ureq(my_url, context=ssl_context)
page_html = openFile.read()
openFile.close()
```

```python
page_soup = soup(page_html, "html.parser")
page_soup
```

```python
containers = page_soup.findAll("div", {"class" : "_1YokD2 _3Mn1Gg"})
print(len(containers))
print(soup.prettify(containers[0]))
```

```python
container = containers[0]
content = container.findAll("div", {"class", "_1AtVbE col-12-12"})
print(len(content))
```

```python
product_images = []
product_names = []
product_descriptions = []
product_prices = []
product_discount = []
product_deliveryCost = []


for contents in content:
    for product in contents:
        
        image = product.findAll("img", {"class" : "_2r_T1I"})
        if image:
            product_images.append(image[0]["src"])
        else:
            product_images.append(np.nan)
            
        name = product.findAll("div", {"class": "_2WkVRV"})
        if name:
            product_names.append(name[0].text)
        else:
            product_names.append(np.nan)
            
        description = product.findAll("a", {"class": "IRpwTa"})
        if description:
            product_descriptions.append(description[0].text)
        else:
            product_descriptions.append(np.nan)
        
        price = product.findAll("div", {"class": "_30jeq3"})
        if name:
            product_prices.append(price[0].text)
        else:
            product_prices.append(np.nan)
            
        discount = product.findAll("div", {"class": "_3Ay6Sb"})
        if discount:
            product_discount.append(discount[0].text)
        else:
            product_discount.append(np.nan)
            
        delivery_cost = product.findAll("div", {"class": "_2Tpdn3"})
        if delivery_cost:
            product_deliveryCost.append(delivery_cost[0].text)
        else:
            product_deliveryCost.append(np.nan)
            
            
            
print(len(product_images))
print(len(product_names))
print(len(product_descriptions))
print(len(product_prices))
print(len(product_deliveryCost))
print(len(product_discount))
```
